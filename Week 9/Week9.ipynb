```python
import arcgis
import pandas as pd
from arcgis.features import SpatialDataFrame 
```

```python
df = pd.read_csv(
    r"C:\Users\eblimitkk\Desktop\GIS5091\advanced_class\Week 9\all_day.csv"
)
```

```python
new_df = df.drop(['nst','gap','dmin','net'],axis=1)
```

```python
new_df.head()
```

```python
locations = new_df.iloc[:,1:3].values.tolist()
```

```python
locations
```

```python
new_sdf = SpatialDataFrame(new_df,geometry=[arcgis.geometry.Geometry({'x':r[1], 'y':r[0], 
                    'spatialReference':{'wkid':4326}}) for r in locations])
```

```python
new_sdf.head()
```

```python
new_sdf=new_sdf[new_sdf.SHAPE.notna()].copy()
```

```python
new_sdf.reset_index(inplace=True,drop=True)
```

```python
new_list = []
```

```python
new_sdf_low = new_sdf[new_sdf['mag'] < 2.5]
new_list.append(new_sdf_low)
```

```python
new_sdf_lowmid = new_sdf[(new_sdf['mag']>2.5) & (new_sdf['mag'] < 4)]
new_list.append(new_sdf_lowmid)
```

```python
new_sdf_mid = new_sdf[(new_sdf['mag']>4) & (new_sdf['mag'] < 5)]
new_list.append(new_sdf_mid)
```

```python
new_sdf_high = new_sdf[new_sdf['mag'] > 5]
new_list.append(new_sdf_high)
```

```python
for i in new_list:
    i.reset_index(inplace=True, drop=True)
```

```python
new_sdf_high
```

```python
from arcgis.gis import GIS
import os
```

```python
gis = GIS("http://slustl.maps.arcgis.com/home", "kevineblimit_slustl")
```

```python
classed_size_renderer = { "renderer": "ClassedSizeRenderer", "field_name":"mag"}
```

```python
map1 = gis.map()
map1
```

```python
eq = gis.content.import_data(new_sdf_high, title='Magnitudes Greater than 5')
```

```python
eq
```

```python
map1.add_layer(eq,classed_size_renderer)
map1
```

```python
webmap_item_properties = {'title':'Earthquakes from Past Day, Too Strong',
                         'snippet':'Strong',
                         'tags':['day','mag5','earthquakes']}
```

```python
map1.save(webmap_item_properties)
```
